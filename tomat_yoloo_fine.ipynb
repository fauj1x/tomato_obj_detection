```python
import os
from ultralytics import YOLO

# Load a pretrained YOLOv8n model
model = YOLO('yolov8n.pt')

# Define the path to your data.yaml
data_yaml_path = '/content/drive/MyDrive/dataset/tomat-yoloo-1/data.yaml'

# Train the model
results = model.train(
    data=data_yaml_path,
    epochs=100,
    imgsz=640,
    device='cuda' if torch.cuda.is_available() else 'cpu' # Use GPU if available
)

print("\nâœ… Pelatihan model selesai!")
```

```python
import os
from ultralytics import YOLO

# Load the best trained model
model_path = '/content/drive/MyDrive/dataset/runs/detect/train/weights/best.pt'
model = YOLO(model_path)

# Define the path to the image for prediction
image_path = '/content/drive/MyDrive/dataset/coba.jpg' # Pastikan nama file dan path sudah benar

# Run prediction on the image
results = model.predict(source=image_path, save=True, save_txt=True, conf=0.2)

print("\nâœ… Prediksi selesai! Hasil disimpan di folder 'runs/detect/predict'.")
```
